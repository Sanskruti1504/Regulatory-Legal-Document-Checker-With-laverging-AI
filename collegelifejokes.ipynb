```python
import random

# List of jokes related to college life
college_jokes = [
    "Why did the math book look sad? Because it had too many problems.",
    "Why don’t college students trust stairs? They’re always up to something.",
    "Why was the computer so tired? It had a hard drive.",
    "Why did the student eat his homework? Because the teacher said it was a piece of cake.",
    "What do you call a group of musical college students? A major band!"
]

# Initial messages to simulate a chatbot
messages = [
    {
        "role": "system",
        "content": "You are a helpful bot that provides college life jokes when prompted."
    }
]

while True:
    prompt = input("Enter a prompt related to college life (type 'exit' to quit): ")

    if prompt.lower() == 'exit':
        print("Goodbye! Have a great day!")
        break

    # Append user input to messages
    messages.append(
        {
            "role": "user",
            "content": prompt
        }
    )

    # Randomly select a joke from the list of college jokes
    joke = random.choice(college_jokes)

    # Add the assistant's joke to the conversation
    messages.append(
        {
            "role": "assistant",
            "content": joke
        }
    )

    # Print the joke to the user
    print(f"Assistant: {joke}\n")
```

```python
import os
from dotenv import load_dotenv
import google.generativeai as genai

# Load API key from .env file
load_dotenv()
API_KEY = os.getenv("GEMINI_API_KEY")

# Set the Gemini API key
genai.configure(api_key=API_KEY)

def get_joke():
    """Fetch a joke from the Google Generative AI API."""
    prompt = "Tell me a funny and light-hearted joke."
    try:
        # Generate text using the Gemini API
        response = genai.generate_text(
            model="text-bison-001",  # Specify the Gemini model
            prompt=prompt,
            temperature=0.7,        # Adjust for creativity
            max_output_tokens=60    # Limit the length of the response
        )
        # Extract and return the joke
        return response.candidates[0]["output"]
    except Exception as e:
        return f"Error generating a joke: {e}"

def chatbot():
    """Simple chatbot interface to tell jokes."""
    print("Welcome to the Joke Bot! Type 'joke' for a joke or 'exit' to quit.")
    while True:
        user_input = input("You: ").strip().lower()
        if user_input == "joke":
            joke = get_joke()
            print(f"Joke Bot: {joke}")
        elif user_input == "exit":
            print("Joke Bot: Goodbye! Have a funny day!")
            break
        else:
            print("Joke Bot: Please type 'joke' or 'exit'.")

if __name__ == "__main__":
    chatbot()
```
